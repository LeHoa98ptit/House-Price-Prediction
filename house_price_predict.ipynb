### Processing Missing Data

- First, We need detect mising values. 
- Then, we need  to get rid of them for the next step of our work. 


```python
def missing_percentage(df):
    """A function for returning missing ratios"""
    total = df.isnull().sum().sort_values(ascending=False)
    
    return pd.concat([total, (total/len(df)*100)], axis=1, keys=["Total", "Percent"])[total!=0]
```

```python
missing_percentage(features)
```

```python
# Checking "NaN" values

missing = missing_percentage(features)

plt.figure(figsize=(20, 7))
sns.barplot(x=missing.index, y=missing["Percent"], palette="Reds_r")
plt.xticks(rotation=90)

display(missing.T.style.background_gradient(cmap="Reds", axis=1))
```
